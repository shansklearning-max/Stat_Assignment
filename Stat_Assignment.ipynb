# **Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?**

Ans - In hypothesis testing, we make decisions based on sample data to either reject or fail to reject a null hypothesis (H‚ÇÄ). However, these decisions can sometimes be incorrect, leading to two types of errors:

**1. Type I Error (False Positive)**

***Definition***: Occurs when the null hypothesis (H‚ÇÄ) is true, but we incorrectly reject it.

***Example***: Concluding that a new drug works when, in reality, it does not.

Probability of Type I Error (Œ±): This is also known as the level of significance, often set at 0.05 or 5%.

***Impact***: It leads to false claims or unnecessary actions, such as approving an ineffective product or policy.

**2. Type II Error (False Negative)**

***Definition***: Occurs when the null hypothesis (H‚ÇÄ) is false, but we fail to reject it.

***Example***: Concluding that a new drug does not work when, in fact, it does.

Probability of Type II Error (Œ≤): The complement of test power, i.e., Power = 1 - Œ≤.

***Impact***: It leads to missed opportunities or delayed improvements, such as rejecting a beneficial innovation.

**Impact on Decision-Making**

***Type I Error***: Increases the risk of taking wrong action based on false evidence. Decision-makers may implement ineffective or harmful strategies.

***Type II Error:*** Increases the risk of inaction when change is actually needed. Decision-makers might miss valuable discoveries or fail to solve real problems.

Therefore, in any hypothesis test, a balance between Œ± (risk of Type I error) and Œ≤ (risk of Type II error) must be maintained depending on the situation‚Äôs consequences.

# **Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?**


Ans -

**Definition**:

The P-value (probability value) is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis (H‚ÇÄ) is true.

In simpler terms, it tells us how likely our sample data is, if there were really no effect or no difference (i.e., if H‚ÇÄ is true).

***Interpretation***:

If the P-value ‚â§ Œ± (level of significance)
‚Üí We reject the null hypothesis (H‚ÇÄ).
‚Üí It means the observed data is statistically significant, and the result is unlikely to have occurred by chance.

If the P-value > Œ±
‚Üí We fail to reject the null hypothesis (H‚ÇÄ).
‚Üí It means the data does not provide strong enough evidence against H‚ÇÄ.

***Example***:

Suppose Œ± = 0.05 and the test gives P-value = 0.02.
Since 0.02 < 0.05, we reject H‚ÇÄ and conclude that there is a statistically significant effect or difference.

# **Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?**

Ans -

**1. What is a Confidence Interval -**

A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as the population mean) with a certain level of confidence.

For example:
A 95% confidence interval means that if we repeatedly took samples and calculated intervals, 95% of those intervals would contain the true population parameter.

**2. What is the Margin of Error (ME)?**

The margin of error represents how much the sample estimate (like sample mean) could vary from the true population value.
It is added and subtracted from the sample estimate to create the confidence interval.
# Confidence Interval =ùëãÀâ¬± Margin of Error

3. How Margin of Error Influences the Width of a Confidence Interval

***Larger Margin of Error ‚Üí Wider Confidence Interval***

*   Shows more uncertainty.

*    Indicates that the estimate is less precise.

*    Happens when sample size is small or confidence level is high.

***Smaller Margin of Error ‚Üí Narrower Confidence Interval***

*    Shows more certainty.

*    Indicates high precision in the estimate.

*    Happens with larger sample sizes or lower confidence levels.

**4. Interpretation:**

A wider confidence interval means the estimate is less precise, while a narrower interval means greater precision.
However, increasing precision (narrow interval) usually means reducing confidence level or increasing sample size.

Example:

If the 95% CI for average weight is:
(68 kg, 74 kg)
This means we are 95% confident that the true population mean lies between 68 and 74.

If margin of error increases, the interval might become (66 kg, 76 kg) ‚Äî wider and less precise.


**# Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?**

Ans -

An ANOVA (Analysis of Variance) test is used to determine whether there are statistically significant differences between the means of three or more groups. Its purpose is to extend hypothesis testing beyond two groups by comparing the variation between group means to the variation within the groups. If the between-group variation is significantly larger, ANOVA concludes that at least one group mean is different. ANOVA relies on key assumptions: the data in each group should be approximately normally distributed, the groups should have equal variances (homogeneity of variance), and observations must be independent. By testing all group means simultaneously, ANOVA avoids the increased error risk that would occur if multiple t-tests were performed separately.

# **Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.**


```python
import numpy as np
from scipy.stats import norm

# Sample data
data = [52, 48, 50, 51, 49, 53, 47, 50, 52, 51]

# Known population mean and standard deviation
mu = 50            # population mean
sigma = 2          # population standard deviation

# Calculate sample statistics
sample_mean = np.mean(data)
n = len(data)

# Z-test statistic
z_score = (sample_mean - mu) / (sigma / np.sqrt(n))

# P-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)
```

# **Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate non-normal distribution: exponential
population = np.random.exponential(scale=1, size=100000)

# Draw multiple samples and compute sample means
sample_means = []
n = 50  # sample size
num_samples = 1000

for _ in range(num_samples):
    sample = np.random.choice(population, size=n, replace=True)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means (CLT demonstration)
plt.hist(sample_means, bins=30)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem Demonstration\n(Sample Means of Exponential Distribution, n=50)")
plt.show()
```

# **Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    # Convert to NumPy array
    data = np.array(data)

    # Sample statistics
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    n = len(data)

    # Z-value for confidence level
    z_value = norm.ppf((1 + confidence) / 2)

    # Margin of Error
    margin_error = z_value * (std / np.sqrt(n))

    # Confidence Interval
    lower = mean - margin_error
    upper = mean + margin_error

    print(f"Sample Mean: {mean}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower}, {upper})")

    # Plotting
    plt.axvline(mean, linestyle="--")
    plt.axvline(lower, linestyle="--")
    plt.axvline(upper, linestyle="--")

    plt.hist(data, bins=15)
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.show()

    return lower, upper
```

```python
sample_data = [12, 15, 14, 10, 18, 16, 17, 13, 11, 15]
plot_confidence_interval(sample_data)
```

# **Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.**


```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([25, 30, 20, 25])

# Expected frequencies (must have same total count)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
```

Based on the Chi-square goodness-of-fit test, we compare the observed frequencies with the expected ones. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the observed distribution does not differ significantly from the expected distribution. If the p-value is less than 0.05, we reject the null hypothesis and conclude that the observed data deviates significantly from the expected pattern.
